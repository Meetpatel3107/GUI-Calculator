```python
from tkinter import*
import math

def click(value):
    
     ex=entryfield.get()
     answer=""

     if value =='⬅':  
       ex=ex[0:len(ex)-1]
       entryfield.delete(0,END)
       entryfield.insert(0,ex)
       return
    
     elif value =='AC':
        entryfield.delete(0,END)

     elif value =='√':
         answer=math.sqrt(eval(ex))

         entryfield.delete(0,END)
         entryfield.insert(0,answer)
     
     elif value == 'π':
        answer=math.pi

        entryfield.delete(0,END)
        entryfield.insert(0,answer)
     
     elif value == 'cosθ':
        answer=math.cos(math.radians(eval(ex)))
        
        entryfield.delete(0,END)
        entryfield.insert(0,answer)
     
     elif value == 'tanθ':
        answer=math.tan(math.radians(eval(ex)))
        
        entryfield.delete(0,END)
        entryfield.insert(0,answer)
     
     elif value == 'sinθ':
        answer=math.sin(math.radians(eval(ex)))

     elif value == '2π':
        answer=2*math.pi

     elif value == 'cosh':
        answer=math.cosh(eval(ex))

     elif value == 'tanh':
        answer=math.tanh(eval(ex))
      
     elif value == 'sinh':
        answer=math.sinh(eval(ex))

     elif value == '∛':
        answer=eval(ex)**(1/3)

     elif value == 'x^y':
        entryfield.insert(END,'**')
        return
     
     elif value == 'x\u00B3':
        answer = eval(ex)** 3

     elif value == 'x\u00B2':
        answer = eval(ex)** 2

     elif value == 'ln':
        answer = math.log2(eval(ex))

     elif value == 'deg':
        answer = math.degrees(eval(ex))

     elif value == 'rad':
        answer = math.radians(eval(ex))

     elif value == 'e':
        answer = math.e

     elif value == 'log10':
        answer = math.log10(eval(ex))

     elif value == '÷':
        entryfield.insert(END,'/')
        return

     elif value == '=':
        answer=eval(ex)

     else:
        entryfield.insert(END,value)
        return


      
    
     entryfield.delete(0,END)
     entryfield.insert(0,answer)

root= Tk()
root.title("calculator")
root.config(bg="dodgerblue3")
root.geometry("760x340")


entryfield=Entry(root,font=('arial',20,'bold'), bg="dodgerblue3",fg='white', bd=10, relief=SUNKEN,width=40,)
entryfield.grid(row=0,column=0,columnspan=8)


button_text_list = ["⬅", "AC", "√", "+", "π", "cosθ", "tanθ", "sinθ",
                    "1", "2", "3", "-", "2π", "cosh", "tanh", "sinh",
                    "4", "5", "6", "*", '∛', "x^y", "x\u00B3", "x\u00B2",
                    "7", "8", "9", "÷", "ln", "deg", "rad", "e",
                    "0", ".", "%","=","log10","(", ")","x!"]

rowvalue=1
columnvalue=0
for i in button_text_list:
 

 
   if i == '+':
    
      button=Button(root,width=5,height=2,bd=2,text=i, bg='dodgerblue3',fg='black',relief=SUNKEN,
              font=('arial',18,'bold'),activebackground='dodgerblue3',command=lambda button=i:click(button))
      button.grid(row=rowvalue, column=columnvalue, pady=2,padx=2)
    
     
   else:
    button=Button(root,width=5,height=2,bd=2,relief=SUNKEN,text=i, bg='dodgerblue3',fg='#1C1C1C',
              font=('arial',18,'bold'),activebackground='dodgerblue3',command=lambda button=i:click(button))
    button.grid(row=rowvalue,column=columnvalue,pady=2,padx=2)

   columnvalue+=1
   if columnvalue>7:
      rowvalue+=1
      columnvalue=0


root.mainloop()


```
